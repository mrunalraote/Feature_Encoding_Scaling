```python
import pandas as pd

df = pd.read_csv("adult.csv")
df.head()
```

```python
df.info()
```

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['income'] = le.fit_transform(df['income'])
```

```python
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
df[numerical_cols].describe()
```

```python
df.to_csv("adult_processed.csv", index=False)
```

In this task, the Adult Income Dataset was preprocessed using feature encoding and scaling techniques. Categorical features were encoded using Label Encoding (for ordered data) and One-Hot Encoding (for unordered data). Numerical features were scaled using StandardScaler to bring them to a common scale. These steps ensure the dataset is fully prepared for machine learning algorithms and improve model performance.
